### Predicting day-ahead energy consumption using ANN: 
Traditional methods for time series forecasting like ARIMA has its limitation as it can only be used for univariate data and one step forecasting. VAR(vector autoregression) model can do multivariate regression, or, it is observed in various studies that deep learning models like ANN(MLP) or RNN(LSTM) outperform traditional forecasting methods on multivariate time series data.

```python
from datetime import datetime as dt, timedelta
import pandas as pd
import numpy as np
import loadForecast as lf
import mpld3
mpld3.enable_notebook()
%matplotlib inline
```

```python
f = 'data/NORTH.csv'
df = pd.read_csv(f)
all_X = lf.makeUsefulDf(df).to_numpy()
all_y = df['load'].to_numpy()
# convert boolean features into float32 variables
all_X = np.asarray(all_X).astype(np.float32)
# print(all_X.shape, all_y.shape)
# print(all_X, all_y)
```

```python
predictions, accuracy = lf.hour_ahead_predictions(all_X, all_y)
```

```python
print('Percent accuracy (MAPE). Train: {}.  Test: {}'.format(100-accuracy['train'], 100-accuracy['test']))

df_r = pd.DataFrame()
df_r['predicted_load'] = predictions
df_r['actual_load'] = [float(f) for f in all_y[-8760:]]
df_r.index = [dt(2018, 1, 1, 0) + timedelta(hours=1)*i for i in range(8760)]
df_r.plot(figsize=(18, 8), title="{} Texas 2018: prediction v. actual load".format(f[5:10]))
```

The new structure effectively combines 24 hourly models. But instead of calculating a single hour, we combine all weights into one flat, fully-connected dense layer (half of window * features number of nodes). That layer is then fully connected to a 24-hour vector for one day ahead forecast. 

```python
def data_transform(data, timesteps, var='x'):
    m = []  
    for i in range(data.shape[0]-timesteps):  # starting index of a day sample
        m.append(data[i:i+timesteps].tolist())
    if var == 'x':
        t = np.zeros((len(m), len(m[0]), len(m[0][0])))
        for i, x in enumerate(m):  # x is each day sample
            for j, y in enumerate(x):  # y is each hour sample
                for k, z in enumerate(y):  # z is each feature
                    t[i, j, k] = z
    else:
        t = np.zeros((len(m), len(m[0])))
        for i, x in enumerate(m):
            for j, y in enumerate(x):
                t[i, j] = y
    return t
```

```python
window = 24
all_y_rnn = data_transform(all_y, window, var='y')
all_X_rnn = data_transform(all_X, window, var='x')
#print(all_X_rnn.shape, all_y_rnn.shape)
#print(all_X_rnn, all_y_rnn)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow import keras

s = all_X.shape[1]
model_dayAhead = Sequential()
model_dayAhead.add(Dense(s, activation="relu", input_shape=(window, s)))
model_dayAhead.add(Dense(s, activation="relu"))
model_dayAhead.add(Dense(s, activation="relu"))
model_dayAhead.add(Dense(s, activation="relu"))
model_dayAhead.add(Dense(s, activation="relu"))
model_dayAhead.add(Flatten())
model_dayAhead.add(Dense(s*window//2, activation="relu"))
model_dayAhead.add(Dense(window))

nadam = keras.optimizers.Nadam(learning_rate=0.002, beta_1=0.9, beta_2=0.999)
model_dayAhead.compile(optimizer=nadam, loss='mape', metrics=["mean_absolute_error", "mape"])
```

```python
early_stop = keras.callbacks.EarlyStopping(monitor="mean_absolute_error", patience=20)
X_train_rnn, y_train_rnn = all_X_rnn[:-8760, :, :], all_y_rnn[:-8760, :]
X_test_rnn, y_test_rnn = all_X_rnn[-8760:, :, :], all_y_rnn[-8760:, :]
EPOCHS=10
```

```python
# train our NN model
history = model_dayAhead.fit(
    X_train_rnn,
    y_train_rnn,
    epochs=EPOCHS,
    verbose=1,
    callbacks=[early_stop],
)
```

```python
#save the model
model_dayAhead.save('./models/day_ahead_forecastor.h5')
```

```python
from tensorflow.keras.models import load_model
model_dayAhead = load_model('./models/day_ahead_forecastor.h5')
```

```python
predictions = np.array([])
for row in X_test_rnn:
    X_test_rnn_row = row.reshape(1, window, s)
    yhat = model_dayAhead.predict(X_test_rnn_row, verbose=0)
    # print(yhat)
    predictions = np.append(predictions, yhat)

train = np.array([])
for row in X_train_rnn:
    X_train_rnn_row = row.reshape(1, window, s)
    ytrainhat = model_dayAhead.predict(X_train_rnn_row, verbose=0)
    # print(ytrainhat)
    train = np.append(train, ytrainhat)
print(predictions, train)
print(predictions.shape, train.shape)
print(y_test_rnn.shape, y_train_rnn.shape)
```

```python
y_test_rnn_flatten = y_test_rnn.flatten()
y_train_rnn_flatten = y_train_rnn.flatten()
print(y_test_rnn_flatten.shape, y_train_rnn_flatten.shape)
print(y_test_rnn_flatten, y_train_rnn_flatten)
```

```python
def MAPE(predicted, true):
    assert len(predicted) == len(true)
    return sum([abs(x-y)/(y+1e-5) for x, y in zip(predicted, true)])/len(true)*100
```

```python
accuracy = {
    'test': MAPE(predictions, y_test_rnn_flatten),
    'train': MAPE(train, y_train_rnn_flatten)
}
print('Percent accuracy (MAPE). Train: {}.  Test: {}'.format(100-accuracy['train'], 100-accuracy['test']))
```
